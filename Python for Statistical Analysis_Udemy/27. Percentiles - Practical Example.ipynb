## Leveraging Cumulative Density Function

```python
ps=st.norm.cdf(np.linspace(-4,4,30))*100


x_p=np.percentile(data, ps)

x=np.sort(data)
y=np.linspace(0,1,len(data))
plt.plot(x,y*100,label="ECDF")
plt.plot(x_p,ps,label="Percentiles",marker=".",ms=10)
plt.legend()
plt.ylabel("Percentile")
```

```python
ps=st.norm.cdf(np.linspace(-3,3,50))*100
ps=np.insert(ps,0,0)
ps=np.insert(ps,-1,100)
x_p=np.percentile(data, ps)

x=np.sort(data)
y=np.linspace(0,1,len(data))
plt.plot(x,y*100,label="ECDF")
plt.plot(x_p,ps,label="Percentiles",marker=".",ms=10)
plt.legend()
plt.ylabel("Percentile")
```

```python
from scipy.interpolate import interp1d

n=int(1e6)
u=np.random.uniform(size=n)

samp_percentile_1=interp1d(ps/100,x_p,kind="linear")(u)

_,bins,_=plt.hist(data,bins=50,density=True,alpha=0.3,label="Data")
plt.hist(samp_percentile_1,bins=bins,density=True,label="Percentiles",histtype="step")
plt.ylabel("Probability")
plt.legend()
```
